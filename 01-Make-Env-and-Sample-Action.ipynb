## Initializing Environment

```python
# require libraries
import gym
import time 
import pygame
import numpy as np
from IPython import display
import matplotlib.pyplot as plt
```

```python
env = gym.make('FrozenLake-v1', desc=None, map_name='4x4', is_slippery=False)
```

```python
state = env.reset()
plt.axis('off')
plt.imshow(env.render(mode='rgb_array'));
print(f'The new episode will start in state {state}')
```

```python
# environment informations
print(f'number of possible actions are {env.action_space.n} and number of states are {env.observation_space.n}')
```

## Make Sample Action

```python
# 0: Left, 1: Down, 2: Right, 3: Up
action = 1
next_state, reward, done, info = env.step(action)
print(f"After moving down 1 row, the agent is in state: {next_state}")
print(f"After moving down 1 row, we got a reward of: {reward}")
print("After moving down 1 row, the task is", "" if done else "not", "finished")
```

```python
plt.axis('off')
plt.imshow(env.render(mode='rgb_array'));
```

## Trajectory and Episode

```python
state = env.reset()
trajectory = []

for _ in range(5):
    time.sleep(0.1)
    action = env.action_space.sample()
    next_state, reward, done, _ = env.step(action)
    trajectory.append([state, action, next_state, reward, done])
    state = next_state
    
    plt.axis('off')
    plt.imshow(env.render(mode='rgb_array'))
    display.display(plt.gcf())    
    display.clear_output(wait=True)
    
env.close()

print(f'You have just done a Trajectoty:\n{trajectory}')
```

```python
state = env.reset()
done = False
episode = []
t = 0

while not done:
    time.sleep(0.1)
    action = env.action_space.sample()
    next_state, reward, done, _ = env.step(action)
    episode.append([next_state, reward, done])
    state = next_state
    t += 1
    
    plt.axis('off')
    plt.imshow(env.render(mode='rgb_array'))
    display.display(plt.gcf())    
    display.clear_output(wait=True)
    
env.close()

print(f'You achievd {reward} reward in {t} moves in 1 episode:{episode}')
```
